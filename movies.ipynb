```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("movies.csv")
df.head()
```

```python
genre_counts = df['Genre'].value_counts()
print("Most Common Genres:")
print(genre_counts)

```

```python
avg_ratings = df.groupby('Genre')['Rating'].mean().sort_values(ascending=False)
print("\nAverage Rating per Genre:")
print(avg_ratings)

```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8,5))
sns.barplot(x=avg_ratings.index, y=avg_ratings.values, palette="coolwarm")
plt.title("Average Movie Rating per Genre")
plt.xlabel("Genre")
plt.ylabel("Average Rating")
plt.show()
plt.figure(figsize=(6,6))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Most Common Genres")
plt.show()
```
